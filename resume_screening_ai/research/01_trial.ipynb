```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)  
    text = " ".join([word for word in text.split() if word not in stop_words])
    return text
```

```python
def extract_skills_from_cleaned_text(cleaned_text, skills_list):
    found_skills = []
    for skill in skills_list:
        if skill in cleaned_text:
            found_skills.append(skill)
    return found_skills

data['extracted_skills'] = data['cleaned_text'].apply(lambda x: extract_skills_from_cleaned_text(x, skills_list))
```

```python
import re
import pandas as pd

class ResumeProcessor:
    def __init__(self, skills_list=None):
        self.skills_list = skills_list or []

    def extract_experience_in_months(self, text):
        text = text.lower()

        # Extract years
        year_matches = re.findall(r'(\d+(?:\.\d+)?)\s*(?:\+)?\s*(?:years?|yrs?|yr)', text)
        total_years = sum(float(y) for y in year_matches)

        # Extract months
        month_matches = re.findall(r'(\d+(?:\.\d+)?)\s*(?:months?|mos?)', text)
        total_months = sum(float(m) for m in month_matches)

        return round(total_years * 12 + total_months)

    def extract_skills(self, text):
        found_skills = []
        text = text.lower()
        for skill in self.skills_list:
            if skill.lower() in text:
                found_skills.append(skill)
        return list(set(found_skills))  # Remove duplicates

    def process_dataframe(self, df, text_column='resume_text'):
        df['experience_months'] = df[text_column].apply(self.extract_experience_in_months)
        if self.skills_list:
            df['skills_found'] = df[text_column].apply(self.extract_skills)
        return df

    def one_hot_encode_skills(self, df, skills_column='skills_found'):
        for skill in self.skills_list:
            df[f'skill_{skill.lower()}'] = df[skills_column].apply(lambda x: int(skill in x))
        return df
```

```python
role_skill_map = (
    data.groupby('Category')['extracted_skills']
    .sum()  # combine lists
    .apply(lambda x: list(set(x)))  # remove duplicates
    .to_dict()
)
```

```python
# Step 1: Define new function to return both count and percentage
def get_matching_skills(row):
    role_skills = set(role_skill_map.get(row['Category'], []))
    resume_skills = set(row['extracted_skills'])
    match_count = len(role_skills.intersection(resume_skills))
    total_role_skills = len(role_skills)
    match_percent = (match_count / total_role_skills * 100) if total_role_skills > 0 else 0
    return pd.Series([match_count, round(match_percent, 2)])

# Step 2: Apply and assign to two new columns
data[['matching_skill_count', 'matching_skill_percent']] = data.apply(get_matching_skills, axis=1)

# Step 3: De-fragment the DataFrame to avoid PerformanceWarning
data = data.copy()
```

```python
def compute_resume_score(row, role_skill_map = role_skill_map, max_exp_months=140, skill_weight=0.7, exp_weight=0.3):
    # Extract role-based skills
    role_skills = set(role_skill_map.get(row['Category'], []))
    resume_skills = set(row['extracted_skills'])

    # Skill match
    match_count = len(role_skills.intersection(resume_skills))
    total_role_skills = len(role_skills)
    skill_match_percent = (match_count / total_role_skills * 100) if total_role_skills > 0 else 0

    # Experience score (normalized to 100 scale)
    exp_months = row.get('experience_months', 0)
    exp_score = min(exp_months / max_exp_months * 100, 100)

    # Final score (weighted average)
    final_score = (skill_match_percent * skill_weight) + (exp_score * exp_weight)

    return pd.Series([
        match_count,
        round(skill_match_percent, 2),
        round(exp_score, 2),
        round(final_score, 2)
    ])
```

```python
data[['matching_skill_count', 'matching_skill_percent', 'experience_score', 'resume_score']] = data.apply(
    compute_resume_score,
    axis=1,
    role_skill_map=role_skill_map
)

# Optional: clean memory fragmentation
data = data.copy()
```
